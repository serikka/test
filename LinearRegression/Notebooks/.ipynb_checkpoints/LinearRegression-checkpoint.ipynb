```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Implementing Linear Regression 

- predict: given X predicts y_predicted
- gradient descent update each coefficient
- cost function: MSE 

```python
def predict(X, W, b):
    y_predicted = np.dot(X,W) + b
    return y_predicted
```

```python
def cost_function(y_true, y_predicted):
    cost = np.mean((y_true - y_predicted) ** 2)
    return cost
```

```python
# W : 1xN (N = number of coeff)
# X : 1xm (m = number of rows/records)

def gradient_descent(X, y_true, y_predicted):
    # to calcuate gradients of our cost function we need derivative of the cost function (MSE f(X) = (y_true - y_pred)**2)/m -> derivative f'(X) = 2 (y_true - y_pred)
    # if we were using RMSE for cost function our derivative function would've been different below
    m = len(y_true)
    dw = (1/m) * np.dot(X.T, (y_predicted - y_true))
    db = (1/m) * np.sum(y_predicted - y_true)
    return dw, db

def updated_coeffs(W, b, dw, db, learning_rate=0.5):
    W = W - learning_rate*dw
    b = b - learning_rate*db
    return W,b
```

```python
X = np.array([[1,3], [2,6], [3,9],[4,4],[12,3],[0,0],[-4,4],[44,-2]])
y_true = np.array([4, 8, 12, 8, 15, 0, 0, 42])
num_iterations = 200
learning_rate = 0.001 # Try with different learning rate e.g. 0.01 and you'll see that cost function behaves differently
W = np.array([1.122, 0.43])
b = 6.233
logs = []

for iteration in range(num_iterations):
    # calculate pred values
    y_predicted = predict(X, W, b)
    # calculate cost function
    cost = cost_function(y_true, y_predicted)
    # calcuate gradients of our cost function
    dw, db = gradient_descent(X, y_true, y_predicted)
    # updated W (coeffs) and b
    W, b = updated_coeffs(W,b, dw, db, learning_rate)
    print(f"coefficients: {W, b}")
    print(f"cost: {cost}")
    logs.append(cost)
    
```

```python
logs
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(logs, label='cost') 
plt.xlabel('iter') 
plt.legend() 
```
